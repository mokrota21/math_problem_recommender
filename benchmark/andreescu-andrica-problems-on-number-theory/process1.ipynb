```python
import pprint
with open(path, 'r', encoding="utf-8") as f:
    s = f.read()
print(s)
```

## Split by chapters

```python
import re
chapter_pattern = r"^# (\d+\.)*\d+ .*$"
matches = re.finditer(chapter_pattern, s, re.MULTILINE)
chap_s_ind = [match.start() for match in matches]
```

```python
chapters = []
for i in range(len(chap_s_ind) - 1):
    chapters.append(s[chap_s_ind[i]:chap_s_ind[i + 1]])
chapters.append(s[chap_s_ind[-1]:])
```

## Manual Check

```python
for chapter in chapters:
    print(chapter[:100], "\n----------------------\n\n\n")
```

## Split by problem/solution

```python
problem_pattern = r"^Problem (\d+\.)+.*"
problem_splits = []
for chapter in chapters:
    matches = re.finditer(problem_pattern, chapter, re.MULTILINE)
    problems_inds = [match.start() for match in matches]
    end_match = [len(chapter) - 1]
    problems_inds += end_match
    problems = []
    for i in range(len(problems_inds) - 1):
        problems.append(chapter[problems_inds[i]:problems_inds[i + 1]])
    problem_splits.append(problems)
```

```python
for split in problem_splits:
    for problem in split:
        print(problem[:100], "\n\n...\n\n", problem[-100:], '\n\n---------------------\n\n')
```

```python
len(problem_splits), len(chapters)
```

```python
chapter_name_pattern = r"^# ((\d+\.)*\d+) (.*)$"
problem_no_pattern = r"^Problem ((\d+\.)+)"
```

```python
j = 24
print(problem_splits[0:j])
print(*chapters[0:j], sep='\n\n-----------\n\n')
```

```python
chapters_tree = {}

for chapter in chapters:
    match = re.match(chapter_name_pattern, chapter, re.MULTILINE)
    chapter_no = match.group(1)
    chapter_name = match.group(3)
    chapter_no = chapter_no.split(".")
    current = chapters_tree
    for d in chapter_no:
        if d not in current:
            current[d] = {}
        current = current[d]
    current["name"] = chapter_name.strip()

import pprint
pprint.pprint(chapters_tree)
```

```python
def get_chapter_topic_metadata(chapter_no):
    if not chapter_no:
        return None
    res = []
    chapter_no = chapter_no.split('.')
    current = chapters_tree
    for d in chapter_no:
        current = current[d]
        res.append(current['name'])
    return res
```

```python
df_rows = []
for id, split in enumerate(problem_splits):
    # print(chapters[id])
    # print(re.match(chapter_name_pattern, chapters[id]))
    try:
        chapter_no = re.match(chapter_name_pattern, chapters[id], re.MULTILINE).group(1)
    except Exception as e:
        chapter_no = None
    for problem in split:
        try:
            match = re.match(problem_no_pattern, problem)
            problem = "Problem." + problem[match.end():]
            problem_no = match.group(1)
        except Exception as e:
            problem_no = None

        chapter_meta = get_chapter_topic_metadata(chapter_no)
        topic_desc = ""
        for t in chapter_meta:
            topic_desc += f"->({t})"
        topic_desc = topic_desc.strip('->')

        problem_solution = problem.split('Solution')
        if len(problem_solution) == 1:
            problem_solution.append(None)
        else:
            problem_solution[1] = "Solution" + problem_solution[1]
        
        row = {}
        row["TopicMetadata"] = topic_desc
        row["Problem"] = problem_solution[0]
        row['Solution'] = problem_solution[1]
        row["Problem&Solution"] = problem
        row["Problem Book No"] = problem_no
        
        df_rows.append(row)
```
