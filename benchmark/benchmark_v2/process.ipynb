```python
import json

df = []
with open('doccano_labeled.jsonl', 'r', encoding='utf-8') as f:
    for record in f:
        df.append(json.loads(record))
```

```python
import pandas as pd

df = pd.DataFrame(df)
df
```

```python
df = df.explode('label').reset_index().drop('index', axis=1).dropna()
df
```

```python
def highlight(row):
    data = row['label']
    return row['text'][data[0]:data[1]]

df['text'] = df.apply(highlight, axis=1)
df['label'] = df.apply(lambda x: x['label'][2], axis=1)
df
```

```python
df = df[['text', 'label']].groupby('label').agg(list)
```

```python
df.to_json('ask_model.json')
```
